### Метеослужба предоставила данные о среднемесячной температуре воздуха в г.Москве за каждый месяц 2022 года ( https://world-weather.ru/pogoda/russia/moscow/2022/?ysclid=lhwolo0ibh794996410). Построить столбчатые диаграммы для дня и ночи.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
pog = ['Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь', 'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь']
gor = [f"{i}" for i in pog]

den = [-5, 1, 3, 8, 14, 23, 25, 27, 12, 9, 0, -4]
noch = [-6, -2, -3, 4, 7, 14, 16, 17, 8, 6, -1, -5]

width = 0.3

x = np.arange(len(gor))

fig, ax = plt.subplots(figsize=(15, 10))
rects1 = ax.bar(x - width/2, den, width, label='День', alpha=0.6, color='r')
rects2 = ax.bar(x + width/2, noch, width, label='Ночь', alpha=0.6, color='b')
ax.set_title('Среднемесячная погода в москве за 2022г.')

ax.set_xticks(x)
ax.set_xticklabels(gor)
ax.axhline(y=0, color='k')
ax.legend()

plt.show()
```
