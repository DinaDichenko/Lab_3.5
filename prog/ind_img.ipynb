### Найти какое-либо изображение в сети Интернет. Создать ноутбук, в котором будет отображено выбранное изображение средствами библиотеки matplotlib по URL из сети Интернет.

```python
import matplotlib.pyplot as plt
from PIL import Image
import requests
from io import BytesIO
```

```python
response = requests.get('https://damion.club/uploads/posts/2022-02/1645738281_32-damion-club-p-arti-milikh-kotikov-art-33.jpg')
img = Image.open(BytesIO(response.content))

ax = plt.gca ()
ax.get_xaxis ().set_visible ( False )
ax.get_yaxis ().set_visible ( False )

plt.imshow(img)
plt.show()
```
